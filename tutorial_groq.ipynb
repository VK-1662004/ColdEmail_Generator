```python
from langchain_core.prompts import PromptTemplate
from langchain_core.language_models import LLM  # Assuming `llm` is an instance of your language model

# Define prompt template for extraction
prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from a careers page of a website.
    Your job is to extract the job postings and return them in JSON format containing the following keys: 'role', 'experience', 'skills', and 'description'.
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):
    """
)

# Initialize your language model instance here
# Assuming `llm` is already instantiated

# Function to split text into chunks
def split_text(text, max_chars=4000):
    return [text[i:i + max_chars] for i in range(0, len(text), max_chars)]

# Process each chunk
def extract_job_postings(page_data):
    chunks = split_text(page_data)
    all_results = []
    
    for chunk in chunks:
        # Create the chain with the prompt template and the LLM
        chain_extract = prompt_extract | llm
        try:
            # Invoke the chain with each chunk of page data
            res = chain_extract.invoke(input={'page_data': chunk})
            all_results.append(res.content)
        except Exception as e:
            print(f"Error processing chunk: {e}")
    
    return all_results

# Sample page data
page_data = "Your scraped careers page data here"  # Replace with actual data

# Run the extraction and print the results
job_postings = extract_job_postings(page_data)
print(job_postings)
```
